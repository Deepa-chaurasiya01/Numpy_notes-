# Join & Merge (Creating Dataset)

```python
# Dataframe of number of sales made by an employees
sales = {'Jones': 10000,
         'Chris': 5000,
         'Piyush': 440,
         'Meera': 6700,
         'Rahul': 300
        }

# Dataframe of all employees and the region they work in
region = {'Jones': 'West',
          'Chris': np.nan,
          'Piyush': 'West',
          'Meera': np.nan,
          'Anthony': 'East',
          'Ellen': 'South',
          'Josh': 'West',
          'Simran': 'East',
          'Oscar': 'North',
         }       
```

**Convert dictionary to DataFreames**

```python
sales_df = pd.DataFrame.from_dict(sales, orient = 'index', columns = ['sales'])
region_df = pd.DataFrame.from_dict(region, orient = 'index', columns = ['region'])
```

```python
sales_df
```

```python
region_df
```



---



## Join


Inner Join. : Use intersection of  two sets.


Left Outer Join. : Use keys from left object

Right Outer Join. : Use keys from right object

Full Outer Join or simply Outer Join. : Use union of keys


```python
joined_df = region_df.join(sales_df, how = 'left')
joined_df
```

```python
joined_df = region_df.join(sales_df, how = 'right')
joined_df
```

```python
joined_df = region_df.join(sales_df, how = 'inner')
joined_df
```

```python
joined_df = region_df.join(sales_df, how = 'outer')
joined_df
```

# Merge

```python
region_df.index.name = 'names'
sales_df.index.name = 'names'
```

```python
pd.merge(region_df, sales_df, on = 'names')
```

```python
pd.merge(region_df, sales_df, on = 'names', how = 'left')
```

```python
pd.merge(region_df, sales_df, on = 'names', how = 'right')
```

```python
pd.merge(region_df, sales_df, on = 'names', how = 'inner')
```

```python
pd.merge(region_df, sales_df, on = 'names', how = 'outer')
```
